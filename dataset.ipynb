```python
import pandas as pd
import re
```

```python
internet_prices_by_country_df = pd.read_html("https://www.numbeo.com/cost-of-living/country_price_rankings?displayCurrency=USD&itemId=33")[1]
internet_prices_by_country_df = internet_prices_by_country_df.drop(columns=[0, 2])
internet_prices_by_country_df.columns = ['Country', 'Internet Price (USD)']
internet_prices_by_country_df['Internet Price (USD)'] = internet_prices_by_country_df['Internet Price (USD)'].apply(lambda price: re.sub('[^0-9\.]', '', price))
internet_prices_by_country_df['Internet Price (USD)'] = internet_prices_by_country_df['Internet Price (USD)'].apply(lambda price: float(price))

internet_prices_by_country_df
```

```python
salaries_by_country_df = pd.read_html("https://www.numbeo.com/cost-of-living/prices_by_country.jsp?displayCurrency=USD&itemId=105")[1]
salaries_by_country_df = salaries_by_country_df.drop(columns=['Rank'])
salaries_by_country_df['Average Monthly Net Salary (After Tax)'] = salaries_by_country_df['Average Monthly Net Salary (After Tax)'].apply(lambda salary: float(salary))
salaries_by_country_df
```

```python
salaries_and_internet_df = salaries_by_country_df.merge(right=internet_prices_by_country_df, on="Country")
salaries_and_internet_df['Percentage of Internet on Income (%)'] = (salaries_and_internet_df['Internet Price (USD)'] / salaries_and_internet_df['Average Monthly Net Salary (After Tax)']) * 100.0
salaries_and_internet_df = salaries_and_internet_df.sort_values(by="Percentage of Internet on Income (%)", ascending=False)
salaries_and_internet_df
```

```python
salaries_and_internet_df.to_csv(path_or_buf="somehting.csv")
```
